```python
import numpy as np
import pandas as pd
```

```python
from google.colab import files
uploaded=files.upload()
df=pd.read_csv("/content/Car.csv")
df
```

```python
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df.head(5)
```

```python
#1
X = df.drop("price", axis=1)
Y = df["price"]
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
#2
categorical_columns = X.select_dtypes(include=["object"]).columns
print("Categorical Columns:")
print(categorical_columns)
X = pd.get_dummies(X, columns=categorical_columns).astype(int)
print("\nFirst 5 rows after encoding:")
print(X.head())
```

```python
#3
numerical_columns = ["year", "mileage", "tax", "mpg", "engineSize"]
scaler = StandardScaler()
X[numerical_columns] = scaler.fit_transform(X[numerical_columns])
print("\nFirst 5 rows after scaling:")
print(X.head())
```

```python
#4
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33,random_state=42)
print("\nShapes")
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
```

```python
#5
model = LinearRegression()
model.fit(X_train, y_train)
print("\nIntercept:")
print(model.intercept_)
print("\nCoefficients:")
print(model.coef_)
```

```python
#6
y_pred = model.predict(X_test)
print("\nFirst 10 Predictions")
print(y_pred[:10])
print("\nFirst 10 Actual Values")
print(y_test.iloc[:10].values)
```

```python
#7
r2 = r2_score(y_test, y_pred)
print("\nR² Score:", r2)
print("\nInterpretation:")
if r2 >= 0.90:
    print("Excellent model performance.")
elif r2 >= 0.75:
    print("Good model performance.")
elif r2 >= 0.50:
    print("Moderate model performance.")
else:
    print("Poor model performance.")
```

```python
#8
joblib.dump(model, "LR_ford_car.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(list(X.columns), "columns.pkl")
print("\nModel and preprocessing files saved successfully.")
```
